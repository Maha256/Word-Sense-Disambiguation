```python
R_patterns = [(r'won\'t','will not'),
             (r'can\'t','cannot'),
             (r'i\'m','i am'),
             (r'(\w+)\'ll','\g<1> will'),
             (r'(\w+)n\'t','\g<1> not'),
             (r'(\w+)\'ve','\g<1> have'),
              (r'(\w+)\'s','\g<1> is'),
             (r'(\w+)\'re','\g<1> are'),
             ]
class REReplacer(object):
    def __init__(self,patterns=R_patterns):
        self.patterns = [(re.compile(regex),rep1) for (regex,rep1) in patterns]
    def replace(self,text):
        s=text
        for(pattern,rep1) in self.patterns:
            s=re.sub(pattern,rep1,s)
        return s
```

```python
rep_word=REReplacer()
x1=[]
for i in x:
    n=rep_word.replace(i)
    n=n.translate(str.maketrans("","",string.punctuation))
    x1.append(n)
```

```python
def pb(word,sense):
  cnt=0
  for i in range(0,len(words)):
      for w in words[i]:
          if(word==w and y[i]==sense):
              cnt+=1
  if(sense=="Financial Institution"):
      return((cnt+1)/(cnt_f+voc_len))
  else:
      return((cnt+1)/(cnt_r+voc_len))
```

```python
for i in x_test:
    sent=nltk.word_tokenize(i)
    pb_f=prior_f
    pb_r=prior_r
    for s in sent:
        #print(s)
        pb_f+=math.log2(pb(s,"Financial Institution"))
        pb_r+=math.log2(pb(s,"River Border"))
    print(pb_f,pb_r,i)
    if(pb_f > pb_r):
        print("Finance")
    else:
        print("River")
```
